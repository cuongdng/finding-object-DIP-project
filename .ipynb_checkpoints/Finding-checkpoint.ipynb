# Import Modules

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import imutils
import time
plt.rcParams["figure.figsize"] = (10,8) # set size for plt figure
```

# Read Image

```python
image = '15' #<----- Choose which image to execute: 12 or 15
img = cv2.imread("image" + image + "_half.jpg")
#defind list of template
if image == '12':
    img_list = ["strawberry", "ice_cream", "butterfly", "bone", "bow", "hotdog", "cloud", "ghost", "ball", "cheese", "cock", "beetle"] #<----Using this for 12 object image----
else:
    if image == '15':
        img_list = ["balloon", "bow", "cake", "car", "duck", "grapes", "ice_cream", "rabbit", "sailboat", "straw_berry", "teddy_bear", "tennis_ball", "toy", "unicorn", "water_melon"]  #<---- using this for 15 object image-----
```

# Auto Extract Object From Original Image (Optional)

```python
img_object = cv2.imread("image" + image + "_object.jpg")
original = img_object.copy()
gray = cv2.cvtColor(img_object, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray, (3,3), 0)
canny = cv2.Canny(blurred, 120, 255, 1)
kernel = np.ones((5,5),np.uint8)
dilate = cv2.dilate(canny, kernel, iterations=1)
# Find contours
cnts = cv2.findContours(dilate, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cnts = cnts[0] if len(cnts) == 2 else cnts[1]

image_number = 0
for c in cnts:
    x,y,w,h = cv2.boundingRect(c)
    cv2.rectangle(img_object, (x, y), (x + w, y + h), (36,255,12), 2)
    ROI = original[y:y+h, x:x+w]
    cv2.imwrite("object_extract_"+ image + "/ROI_{}.png".format(image_number), ROI)
    image_number += 1
```

# Main Program - Loop Matching With All Template

```python
start_time = time.time()
img_copy = img.copy() #<--- Draw Bounded box on the copy, so it won't affect to the image when we re compile the code
for i in img_list:
    tem = cv2.imread("object_extract_" + image + "/"+ i + ".png") # <--change the dir location depend on which image
    tem = cv2.Canny(tem, 50, 200)
    #loop over scales of template
    found = None
    # after running so so so many time, the ideal set linspace is (0.5, 0.75, 3)
    if image == "15": 
        start_range = 0.5
        end_range = 0.7
    else: 
        start_range = 0.65
        end_range = 0.75
    for scale in np.linspace(start_range, end_range, 3)[::-1]:
        resized = imutils.resize(tem, width=int(tem.shape[1] * scale))
        edged = cv2.Canny(img, 50, 200)
        result = cv2.matchTemplate(edged, resized, cv2.TM_CCOEFF_NORMED)

        (_, maxVal, _, maxLoc) = cv2.minMaxLoc(result)
        
        if found is None or maxVal > found[0]:
            found = (maxVal, maxLoc)
            h, w = resized.shape[:2]
    (_, maxLoc) = found
    cv2.putText(img_copy, i, (maxLoc[0], maxLoc[1]), cv2.FONT_HERSHEY_SIMPLEX, 2, (255,0,0), 6, True) #display template's name aka object name, no deep learning here :D
    cv2.rectangle(img_copy, (maxLoc[0], maxLoc[1]), (maxLoc[0] + w, maxLoc[1] + h), (0, 0, 255), 3) #draw Bounded box
print("Execution time:", time.time() - start_time)
```
